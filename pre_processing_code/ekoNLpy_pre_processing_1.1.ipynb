```python
from ekonlpy.sentiment import MPCK
from ekonlpy.tag import Mecab
import numpy as np
import pandas as pd
import csv
from ekonlpy.tag import Mecab

def multi_pre_processing(df, doc_name_str, max_iterate) : 
    test_list = [[1,2,3,4,5,6]]
    df_pre_processing = pd.DataFrame(test_list,columns =['date', 'title', 'content','tokens', 'ngrams', 'callLabel'])
    
    mecab = Mecab()
    mpck = MPCK()
    for i in range(max_iterate+1) :
        print('{}/{}'.format(i, max_iterate))
        text = str(df['내용'][i])
        date = ('20'+str(df['날짜'][i])).replace('.','-')
        title = str(df['제목'][i])

        tokens = mpck.tokenize(text)

        text = str(df['내용'][i]).replace('\n', '')
    # #멀티프로세싱해서 돌리기

        ngrams = mpck.ngramize(tokens)
        df_pre_processing.loc[i] = [date, title, text, tokens, ngrams, 0]

    df_pre_processing.to_csv('{}.csv'.format(doc_name_str)) #맞게 고치기

```

```python
multi_pre_processing(df, 'test', 10)
```
