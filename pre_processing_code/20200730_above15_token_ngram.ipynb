```python
# 의사록은 ngram과 token이 합쳐져 있었기 때문에 분리하여 다른 데이터프레임과 형식 통일하는 코드 작성.
minutes['token'] = ''

# 세미콜론 포함되면 ngram, 없으면 token으로
for idx, row in minutes.iterrows():
    token_lst = []
    ngram_lst = []
    for lst in row['ngram'].split('@@@'):
        token_lst += [token for token in lst.split(',') if not ';' in token]
        ngram_lst += [ngram for ngram in lst.split(',') if ';' in ngram]
    minutes['token'][idx] = token_lst
    minutes['ngram'][idx] = ngram_lst

minutes.rename(columns={"token":"tokens", "ngram":'ngrams'}, inplace = True)
minutes
```

## 15회 이상 출현한 n-gram과 token 목록 추출

```python
# 각 데이터프레임에서 n-gram과 token 컬럼만 추출
new_bond = bond[['tokens', 'ngrams']]
new_yh = yh[['tokens', 'ngrams']]
new_yhi = yhi[['tokens', 'ngrams']]
new_eday = eday[['tokens', 'ngrams']]
new_minutes = minutes[['tokens', 'ngrams']]
```

```python
# 모든 데이터프레임 통합
concat_df = pd.concat([new_bond, new_yh, new_yhi, new_eday, new_minutes])
concat_df.reset_index(drop=True, inplace=True)
```

```python
concat_df
```

```python
# csv 파일로 저장
concat_df.to_csv('all_token_ngram.csv')
```

```python
# 15회 이상 출현 리스트 작성 함수
import collections

def count_ngram(df_contents) : #dataframe형식으로    //token이랑 ngram 넣어야하니까 함수 두번 돌리세요
    
    sum_content =''
    contents = list(df_contents)
    valid_ngram_list = []
    for content in contents:
        if type(content) == str : #nan이 아니면
            sum_content += content
    
    sum_content = sum_content.replace('[',"")
    sum_content = sum_content.replace(']',"")
    sum_content = sum_content.replace('\'',"")
    sum_content= sum_content.split(', ')
    print('전처리 완료')
    sum_content = list(sum_content)
    
    counter_numbers = collections.Counter(sum_content)
    
    for i in counter_numbers.keys():
        if counter_numbers[i] >= 15:
            valid_ngram_list.append(i)
        
    print('모든 과정 완료')
    return valid_ngram_list
```

```python
# 15회 이상 추출하여 데이터프레임 만들고 csv 파일로 저장
import csv
token_list = count_ngram(concat_df['tokens'])
ngram_list = count_ngram(concat_df['ngrams'])
dic = {}
dic['token_list'] = [token_list]
dic['ngram_list'] = [ngram_list]
tn_df = pd.DataFrame(dic)
tn_df.to_csv(r"C:/NHM/above_15_tokens_ngram.csv", index=False, encoding='utf-8-sig')
```

## 15회 이상 등장한 token과 ngram만 포함하도록 데이터프레임 수정

```python
# 15회 이상 출현 목록 기반으로 컬럼 수정하는 함수
def token_change(token_columns, above_15_lst):
    for i, tokens in enumerate(token_columns):
        tokens = tokens.replace('[',"")
        tokens = tokens.replace(']',"")
        tokens = tokens.replace('\'',"")
        tokens = tokens.split(', ')
        t_list = []
        for j in tokens:
            if j in above_15_lst:
                t_list.append(j)
        token_columns[i] = t_list
```

```python
# 반복 수행하기 위해 리스트 만들기(의사록은 형식이 달라서 따로 작업)
all_lst = [yh, yhi, eday, bond]
```

```python
# token과 ngram에 대해서 15회 이상 출현한 것들만 남도록 수정
for idx, x in enumerate(all_lst):
    token_change(x['tokens'], tn_df['token_list'][0])
    token_change(x['ngrams'], tn_df['ngram_list'][0])
    print('{}번 파일 완료'.format(idx+1))
```

```python
# 의사록은 다른 파일들과 형식이 달라서 따로 작업 수행
def token_change_minutes(token_columns, above_15_lst):
    for i, tokens in enumerate(token_columns):
        t_list = []
        for j in tokens:
            if j in above_15_lst:
                t_list.append(j)
        token_columns[i] = t_list
```

```python
token_change_minutes(minutes['tokens'], tn_df['token_list'][0])
token_change_minutes(minutes['ngrams'], tn_df['ngram_list'][0])
```
