## 전처리 코드 작성 이유  


+ 아무 전처리 없이 바로 텍스트에 token과 ngram 함수 적용하면 '배진우 기자'의 '기자'가 토큰에 포함되는 등의 문제 생김.
+ 기사 내용과 상관없는 부분들은 제거하고 token과 ngram 함수 적용해야 할 필요성이 있음.
+ 따라서 필요없는 부분들 미리 제거하기 위한 코드 작성.

##  각 언론사 데이터프레임 불러오기

```python
# 예시 데이터 프레임들
yhi_df = pd.read_csv('yonhap_in.csv', index_col = 0)
ed_df = pd.read_csv('eday.csv', index_col = 0)
yh_df = pd.read_csv('yonhap.csv', index_col = 0)
```

## 정규표현식 활용한 전처리 코드 활용

+ 각 언론사 텍스트들을 살펴본 후에 제거 대상 목록 선정하였음.  
+ 이데일리에서 제거할 것들  
    - [서울=edaily 배진우 기자]  
    - Copyright ... 배진우 기자 (bigboom@edaily.co.kr)  
+ 연합뉴스, 연합인포맥스에서 제거할 것들  
    - (부산=연합뉴스), (서울=연합인포맥스)  
    - 배진우 기자 =, 배진우 기자=  
    - 하은혜 특파원 =, 조하담 특파원=  
    - goisugn@yna.co.kr
    - (끝)
+ 특수문자 등은 토큰화 과정에서 어차피 제거되기 때문에 제외하고 전처리하였음.

```python
# 각 언론사 내용을 text변수에 저장(숫자는 계속 변경하면서 확인함)
text_ed = ed_df['content'][12]
text_yh = yh_df['content'][12]
text_yhi = yhi_df['content'][12]
```

```python
text_lst = [text_ed, text_yh, text_yhi]
```

```python
# for문 돌려서 세 언론사 모두 전처리 잘 되는지 확인.
for text in text_lst:
    print('전처리 전:\n{}'.format(text))
    print('-'*50)
    # 이데일리 전처리
    text2 = re.sub('\[.*\]', '', text) # [서울=edaily 이민호 기자] 등 제거
    text3 = re.sub('Copyright.+\(?[a-zA-z0-9]+@[a-zA-z.]+\)?', '', text2) # Copyright ... 배진우 기자 (bigboom@edaily.co.kr) 제거

    # 연합뉴스, 연합인포맥스 전처리
    text4 = re.sub('\(\w*=\w*\)', '', text3) # (부산=연합뉴스), (서울=연합인포맥스) 등 제거
    text5 = re.sub('[\w\s]+기자\s?=', '', text4) # 배진우 기자 =, 배진우 기자= 제거
    text6 = re.sub('[\w\s]+특파원\s?=', '', text5) # 하은혜 특파원 =, 조하담 특파원= 제거
    text7 = re.sub('[a-zA-z0-9]+@[a-zA-z.]+', '', text6) # goisugn@yna.co.kr 제거
    text8 = re.sub('\(끝\)', '', text7).strip() # (끝) 제거
    print('전처리 후:\n{}'.format(text8))
    print('-'*150)
```

+ 언론사 별로 if 조건문 활용하면 코드가 복잡해지기 때문에 전처리 코드 통합시킴.
+ 세 언론사 모두에 전처리 코드 돌려본 결과 제거 대상들 모두 처리되는 것 확인.  
(연합뉴스 print() 적용하면 내용이 잘려서 출력되는 문제 있음. text 자체는 문제 없는 것 확인)
