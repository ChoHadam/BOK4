## 모든 지표를 통합한 데이터프레임 만들기

```python
# import
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
bok_rate = pd.read_csv('bok_rate.csv')
bok_rate
```

```python
# 기준금리데이터가 2004년 11월 11일, 2005년 10월 11일 데이터가 있어서 2005년 1월 1일부터의 데이터를 만들어주기위해 20041111부터 시작.
dt_index = pd.date_range(start='20041111', end='20171231')
dt_list = dt_index.strftime("%Y-%m-%d").tolist()

mk_bok_rate = pd.DataFrame(dt_list ,columns = ['date'])
mk_bok_rate
```

```python
mk_bok_rate = pd.merge(mk_bok_rate,bok_rate, on='date',how='left')
mk_bok_rate
```

```python
mk_bok_rate = mk_bok_rate.replace(np.nan, 0)

for i in range(len(mk_bok_rate)):
    if mk_bok_rate['bok_rate'][i] == 0:
        mk_bok_rate['bok_rate'][i] = mk_bok_rate['bok_rate'][i-1]
mk_bok_rate
```

```python
lst = os.listdir()
```

```python
dic_lst = lst[4:]
dic_lst
```

```python
# 각 데이터프레임을 불러와서 한 데이터프레임으로 통합
for x in dic_lst:
    df = pd.read_csv(x)
    df_new = df[['date', 'tone_mkt']]
    if 'dic' in x:
        df_new.rename(columns={"tone_mkt": 'tone_{}'.format(x[17:-4])}, inplace=True)
    mk_bok_rate = pd.merge(mk_bok_rate, df_new, on='date',how='left')
```

```python
mk_bok_rate_drop = mk_bok_rate.dropna()
mk_bok_rate_drop.reset_index(inplace=True, drop=True)
mk_bok_rate_drop.rename(columns={"tone_bond": '보고서', "tone_ex_minutes": '뉴스/보고서',
                                'tone_ex_bond': '뉴스/의사록', 'tone_ex_news': '보고서/의사록',
                                 'tone_minutes': '의사록', 'tone_news': '뉴스'}, inplace=True)
mk_bok_rate_drop = mk_bok_rate_drop[['date', 'bok_rate', 'tone_mkt', '뉴스', '보고서', '의사록', '뉴스/보고서', '뉴스/의사록', '보고서/의사록']]
```

```python
mk_bok_rate_drop
```

```python
mk_bok_rate
```

```python
mk_bok_rate_drop.to_csv('correlation.csv', encoding='utf-8-sig')
mk_bok_rate.to_csv('correlation_for_datastudio.csv', encoding='utf-8-sig')
```

## 상관관계 히트맵 만들기

```python
import seaborn as sns
```

```python
# 한글 글씨체
plt.rc('font',family='Malgun Gothic')
# 마이너스 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

```python
idx = ['bok_rate', 'tone_mkt', '뉴스', '보고서', '의사록', '뉴스/보고서', '뉴스/의사록', '보고서/의사록']
mk_bok_rate_drop = mk_bok_rate_drop.reindex(columns=idx)
```

```python
bok_corr = mk_bok_rate_drop.corr()
```

```python
bok_corr = bok_corr.drop(['bok_rate'])
```

```python
bok_corr
```

```python
plt.figure(figsize=(5,7))
sns.heatmap(bok_corr[['bok_rate']].sort_values(by=['bok_rate'], ascending=False),
            vmin=-0.3, vmax=0.6,
            cmap='coolwarm',
            annot=True);
```
